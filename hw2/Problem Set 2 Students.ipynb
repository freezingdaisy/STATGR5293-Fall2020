```python
# For each grade, calculate the following, with your own script or with the help of this sample notebook:
# Refer to the assignment document for the explicit formulas.

# 1. Average interest rate for loans in that grade
# 2. Realized default rate for loans in that grade
# 3. For non-defaulting loans, the Average Return
# 4. For defaulting loans, compute the average Loss Given Default
# 5. An expected return
# 6. Return of portfolio of given grade
```

```python
# The table has been completed for you. 
# Answer the questions that follow below. 

# Calculate returns
data['return']= data['total_pymnt']/data['loan_amnt'] - 1

# Calculate some statistics conditioned on grade
grades = np.sort(data['grade'].unique())
summary_stats = []
for g in grades:
    temp_data = data[data['grade'] == g] # fill in to get only loans where grade is current grade denoted by variable g
    avg_int_rate = temp_data['int_rate'].mean() # average interest rate for grade
    prevalence = temp_data['default'].mean() # default rate for grade
    avg_rtn_def = temp_data[temp_data['default'] == 1]['return'].mean() # average return if default (default = 1)
    avg_rtn_ndef = temp_data[temp_data['default'] == 0]['return'].mean() # average return if not default (default = 0)
    exp_rtn = avg_rtn_def*prevalence + avg_rtn_ndef*(1 - prevalence) # expected return based on equation E[R] = PD*(Return if Default) + (1 - PD)*(Return if No Default)
    act_rtn_port = temp_data['total_pymnt'].sum() / temp_data['loan_amnt'].sum() -1# expected return of portfolio - total in / total out
    
    summary_stats.append( [avg_int_rate, prevalence, avg_rtn_def, avg_rtn_ndef, exp_rtn, act_rtn_port, avg_rtn_ndef/avg_int_rate] )
    
summary_stats = pd.DataFrame(summary_stats, index=grades, columns=['avg int rate',
                                                                   'default prev', 
                                                                   'avg return default',
                                                                   'avg return no default', 
                                                                   'expected return',
                                                                   'total return portfolio',
                                                                   'ret v ir ratio'
                                                                   ])

display(summary_stats.style.format('{:0.3%}'))
```
